```python
# Writing A Program to Clean A Column Names Votes
def cleanVotes(val):
    val = str(val)
    if val[-1]=='K':
        return int(float(val[0:len(val)-1])*1000)
    elif val[-1]=='M':
        return int(float(val[0:len(val)-1])*1000000)
    else:
        return int(val)
```

```python
df['Votes'] = df['Votes'].map(cleanVotes)
```

```python
# We are goint to write a function to clean Duration Column
def cleanDuration(val):
    val = str(val)
    arr = val.split(' ')
    if len(arr)==2:
        return ( int(arr[0][0:len(arr[0])-1])*60 + int(arr[1][0:len(arr[1])-1]) )
    elif arr[0][-1]=='h':
        return int(arr[0][0:len(arr[0])-1])*60
    elif arr[0][-1]=='m':
        return int(arr[0][0:len(arr[0])-1])
    else:
        return cleanDuration(df['Duration'].mode()[0])
```

```python
df['Duration'] = df['Duration'].map(cleanDuration)
```
